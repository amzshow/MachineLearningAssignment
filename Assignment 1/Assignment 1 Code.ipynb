```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getPrediction(thetha0, thetha1, input):
    return (thetha0 + (thetha1 * input))
```

```python
def getCost(thetha0, thetha1, data):
    sum0 = 0
    n = (float)(len(data))
    for i in range((int)(n)):
        x = data[i][0]
        y = data[i][1]
        prediction = getPrediction(thetha0, thetha1, x)
        sum0 += (prediction - y) ** 2
    
    return sum0 / n
```

```python
def getGradientDescentStepValue(Thetha0, Thetha1, data):
    sum0 = 0
    sum1 = 0
    n = (float)(len(data))
    for i in range((int)(n)):
        x = data[i][0]
        y = data[i][1]
        prediction = getPrediction(Thetha0, Thetha1, x)
        sum0 += (prediction - y)
        sum1 += (prediction - y) * x
    gradient0 = (sum0 / n)
    gradient1 = (sum1 / n)
    return [gradient0, gradient1]
```
